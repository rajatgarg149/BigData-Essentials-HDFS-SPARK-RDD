```python
#! /usr/bin/env python

from pyspark import SparkConf, SparkContext
sc = SparkContext(conf=SparkConf().setAppName("MyApp").setMaster("local[2]"))

import re
```

```python
def parse_article(line):
    try:
        article_id, text = unicode(line.rstrip()).split('\t', 1)
    except ValueError as e:
        return []
    text = re.sub("^\W+|\W+$", "", text, flags=re.UNICODE)
    words = re.split("\W*\s+\W*", text, flags=re.UNICODE)
    word_pair = list(zip(words[:-1], words[1:]))
    word_pair = [str(pair[1].lower()) for pair in word_pair if pair[0].lower() == 'narodnaya']
    return word_pair

wiki = sc.textFile("/data/wiki/en_articles_part/articles-part", 16).flatMap(parse_article).map(lambda x: (x,1)).reduceByKey(lambda x,y: x+y).sortByKey()  
result = wiki.map(lambda x: ('narodnaya_' + str(x[0]), x[1])).collect()

for pair_count in result:
    print "%s\t%d" % (pair_count)
```
